```python
urban = geobr.read_urban_area(year=2015)
```

```python
urban_sjc = urban[urban.code_muni==3549904]
urban_sjc.plot()
```

```python
import matplotlib.pyplot as plt
fig, axis = plt.subplots(figsize=(10,10))

urban_sjc.plot(column= urban_sjc.density, categorical=True, ax= axis, cmap=('RdYlGn'), legend=True)
```

```python
import contextily as cx
```

```python
urban_sjc.crs = {'init': 'epsg:4326'}
census_sjc_2010.crs = {'init': 'epsg:4326'}
```

```python
fig, axis = plt.subplots(figsize=(10,10))
urban_sjc.plot(column= urban_sjc.density, categorical=True, ax= axis, cmap=('RdYlGn'), legend=True)
census_sjc_2010.boundary.plot(ax=axis)

cx.add_basemap(ax=axis, crs=urban_sjc.crs)
```

```python
sjc_lim= census_sjc_2010.dissolve(by='code_muni')
```

```python
fig, axis = plt.subplots(figsize=(10,10))
urban_sjc.plot(column= urban_sjc.density, alpha=0.7, color='darkred', ax=axis)
census_sjc_2010.boundary.plot(ax=axis, color='black', alpha=0.4)
cx.add_basemap(ax=axis, crs=urban_sjc.crs)
axis.tick_params(axis='y', labelrotation=90)
```

```python
from matplotlib_scalebar.scalebar import ScaleBar
from matplotlib.patches import Patch
```

```python
urban_sjc = urban_sjc.to_crs(epsg=31983)
census_sjc_2010= census_sjc_2010.to_crs(epsg=31983)
```

```python
fig, axis = plt.subplots(figsize=(10,10))
urban_sjc.plot(column= urban_sjc.density, alpha=0.7, color='darkred', ax=axis)
census_sjc_2010.boundary.plot(ax=axis, facecolor='black', alpha=0.4, linewidth=0)
cx.add_basemap(ax=axis, crs=urban_sjc.crs)
axis.tick_params(axis='y', labelrotation=90)
axis.add_artist(ScaleBar(dx=1, location='lower right'))
```

```python
# Criando legenda
legend = [Patch(facecolor='black', alpha=0.4,
 label='Limite municipal'),
 Patch(facecolor='darkred', alpha=0.7,
 label='Setores urbanos')
 ]

# Criando o plot editando o código anterior que gerou o mapa
fig, axis = plt.subplots(figsize=(15,10))

# Adicionando título
fig.suptitle("São José dos Campos - SP", fontsize=16)
sjc_lim.plot(ax=axis, color= 'black', alpha=0.4)
urban_sjc.plot(column= urban_sjc.density, alpha=0.7, color='darkred', ax=axis)

# census_sjc_2010.boundary.plot(ax=axis, facecolor='black', alpha=0.4, linewidth=0)
cx.add_basemap(ax=axis, crs=urban_sjc.crs)
axis.tick_params(axis='y', labelrotation=90)

# Adicionando a escala, dx=1 é o padrão para sistemas de coordenadas métricas
# Location indica onde a escala será colocada
axis.add_artist(ScaleBar(dx=1,location='lower right'))
axis.ticklabel_format(useOffset=False, style='plain')

# Definindo limites para a área visualizada
axis.set_ylim(7420001,7477000)
axis.set_xlim(385000,430000)

# Adicionando o N do norte como anotação de texto
axis.text(x=419000, y=7468000, s='N', fontsize=30)

# Adicionando uma seta para indicar o norte
axis.arrow(x=420000, y=7465000, dx=0, dy=12, length_includes_head=True,
 head_width=2000, head_length=1200, overhang=3, facecolor='k')

# Adicione a legenda que personalizamos
axis.legend(handles=legend, loc='upper right', fontsize=16)

# Evite sobreposições no itens do mapa
fig.tight_layout(pad=4)
```
